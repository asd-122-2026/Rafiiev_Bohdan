###**Завдання 1:** Реалізувати процедуру обчислення факторіалу за допомогою циклу for. Оцінити асимптотичну складність алгоритму.

```python
def factorial_iterative(n):
    
    if n < 0:
        return "Помилка: від'ємне число"
    
    result = 1
    # Цикл виконується від 1 до n включно
    for i in range(1, n + 1):
        result *= i
    return result
n = int(input("Введіть n для факторіалу"))
factorial_iterative(n)
```

Асимптотична складність алгоритму обчислення факторіалу через цикл for становить $O(n)$, оскільки кількість операцій множення зростає прямо пропорційно значенню вхідного числа $n$.

###**Завдання 2:** Оцінити асимптотичну складність рекурсивного алгоритму обчислення факторіалу.

```python
def factorial_recursive(n):
    if n <= 1:
        return 1
    return n * factorial_recursive(n - 1)
    
n = int(input("Введіть n для факторіалу"))
print(factorial_recursive(n)) 
```

Рекурсивний алгоритм має часову складність $O(n)$ та просторову складність $O(n)$, оскільки він виконує $n$ послідовних викликів, кожен з яких додає новий фрейм у стек пам'яті.

###**Завдання 3:** Оцінити асимптотичну складність рекурсивного алгоритму обчислення 𝑛-го числа Фібоначчі.

```python
def fibonacci_recursive(n):
    if n <= 1:
        return n
    return fibonacci_recursive(n - 1) + fibonacci_recursive(n - 2)
    
n = int(input("Введіть n для обчислення n-го числа Фібоначчі"))
print(fibonacci_recursive(n))
```

Асимптотична складність алгоритму становить $O(2^n)$, оскільки кількість рекурсивних викликів подвоюється з кожним кроком, утворюючи експоненціальне дерево обчислень.

###**Завдання 4:** Оцінити асимптотичну складність алгоритму сортування злиттям, використовуючи головну теорему рекурсії.

```python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr
        
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])

    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result.extend(left[i:])
    result.extend(right[j:])
    return result
```

Згідно з головною теоремою рекурсії, складність сортування злиттям становить $O(n \log n)$, оскільки задача ділиться на дві підзадачі половинного розміру, а процес їх злиття вимагає лінійного часу.

###Контрольні питання:

- Асимптотична складність у $O$-нотації записується як $O(N^3)$, оскільки при аналізі враховується лише доданок із найвищим степенем.
- Асимптотична складність записується як $O(1.01^N)$, тому що будь-яка експоненціальна функція зростає швидше за будь-яку поліноміальну функцію при великих $N$.
- Асимптотична складність становить $O(N^{1.3})$, оскільки степенева функція має вищий порядок росту порівняно з логарифмічною.
- Ідея полягає в одночасному виконанні незалежних частин задачі на кількох обчислювальних ресурсах для прискорення роботи, що найкраще реалізується в алгоритмі сортування злиттям (Merge Sort).
- Основними способами є апаратне прискорення, розпаралелювання та оптимізація коду, проте найефективнішим є алгоритмічний спосіб, тобто вибір методу з кращою асимптотичною складністю.
