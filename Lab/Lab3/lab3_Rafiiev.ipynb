Сортування методами вставляння:

Найгірший асимптотична складність- O(n^2). Це тому, що кожен новий елемент доводиться вставляти на початок відсортованої частини, зсуваючи всі інші.
Найкраща асимптотична складність - О(n). Це тому, що масив відсортований, і кожен елемент просто порівнюється один раз без зсувів.
Бульбашка:

Найгірший асимптотична складність- O(n^2). Це тому, що кожен елемент багаторазово “просіюється” через масив і порівнюється з сусідніми на кожному проході.
Найкраща асимптотична складність - О(n). Є перевірка обмінів: масив вже відсортований, і алгоритм завершується після одного проходу.
Завдання 2: Реалізувати на Python алгоритм сортування методом бульбашки.

```python
import time

def bubble_sort(arr):
    n = len(arr)

    for i in range(n):
        swapped = False

        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True

        if not swapped:
            break

    return arr


numbers = [5, 2, 9, 1, 5, 6]

start_time = time.time()
print(bubble_sort(numbers))
print("--- %s seconds ---" % (time.time() - start_time))
```
