###**Завдання 1:** Розглянути самостійно Приклад 3 офіційної документації модуля Queue[31]

```python
import queue

def print_n(q, n):
    elements = list(q.queue)
    for i in range(min(n, len(elements))):
        print(elements[i])
```

###**Завдання 2:** Написати функцію print_n(), що друкує елементи черги з його початку до номера n включно

```python
q = queue.Queue()
for item in range(10):
    q.put(item)

print_n(q, 5)
```

###**Завдання 3:** Оцінити асимптотичну складність (у середньому і в найгіршому разі) процедур search, insert і delete роботи з чергою.

1. Insert (put / enqueue):
- Середній та найгірший випадок: $O(1)$
- Додавання в кінець структури відбувається за сталий час, оскільки черга зберігає пряме посилання на останній вузол.
2. Delete (get / dequeue):
- Середній та найгірший випадок: $O(1)$
- Видалення першого елемента не потребує ітерації або переіндексації всієї структури, що забезпечує миттєвий доступ.
3. Search:
- Середній та найгірший випадок: $O(n)$ 
- Для пошуку конкретного значення необхідно послідовно перевірити кожен елемент від початку до кінця черги.

###Контрольні питання:
- **Стек** працює за принципом LIFO (Last In, First Out) — «останній прийшов, першим пішов». Це означає, що доступ до елементів можливий лише з одного боку, який називають вершиною. Кожен новий елемент кладеться поверх попереднього, і щоб дістати найперший доданий об'єкт, необхідно спочатку вилучити всі ті, що були додані пізніше.
- **Черга** базується на принципі FIFO (First In, First Out) — «першим прийшов, першим пішов». У цій структурі додавання елементів відбувається в кінець (хвіст), а видалення — з початку (голови). Це нагадує реальну чергу в магазині, де той, хто став у неї першим, першим отримує обслуговування, а нові люди займають місце позаду.
- **Основна відмінність** між ними полягає в порядку вилучення даних. Стек обертає порядок елементів (як стопка тарілок), тоді як черга зберігає його незмінним. З точки зору реалізації, стек зазвичай простіший у роботі зі стандартними масивами, тоді як для ефективної черги краще використовувати двозв'язні списки або спеціальні буфери, щоб уникнути дорогого зсуву елементів при видаленні з початку.
- **Сфери застосування** цих структур охоплюють майже всі аспекти програмування. Стек незамінний для керування викликами функцій, перевірки парності дужок у коді або реалізації функції скасування останньої дії. Черги ж критично важливі для систем, де завдання мають оброблятися в порядку надходження, наприклад, у планувальниках операційних систем, мережевих буферах або чергах на друк документів.
