###**Завдання 1:** Оцінити асимптотичну складність алгоритму бінарного пошуку в О-нотації.

Складність бінарного пошуку становить $O(\log n)$, оскільки на кожній ітерації простір пошуку скорочується вдвічі.

###**Завдання 2:** Написати функцію, яка б могла повертати датафрейм з таким складом полів: («n», «time»), де n – розмір масиву для функції linear_searh().

```python
import time
import pandas as pd
import numpy as np

def get_linear_search_df(sizes):
    data = []
    for n in sizes:
        arr = list(range(n))
        target = -1
        start = time.perf_counter()
        for i in arr:
            if i == target: break
        end = time.perf_counter()
        data.append({"n": n, "time": end - start})
    return pd.DataFrame(data)
```

###**Завдання 3:** Написати функцію, яка б могла повертати датафрейм з таким складом полів: («n», «time»), де n – розмір масиву для функції bin_searh().

```python
def get_bin_search_df(sizes):
    data = []
    for n in sizes:
        arr = list(range(n))
        target = -1
        start = time.perf_counter()
        low, high = 0, n - 1
        while low <= high:
            mid = (low + high) // 2
            if arr[mid] == target: break
            elif arr[mid] < target: low = mid + 1
            else: high = mid - 1
        end = time.perf_counter()
        data.append({"n": n, "time": end - start})
    return pd.DataFrame(data)
```

###**Завдання 4:** За результатами попередніх завдань побудувати графіки залежності часу виконання процедури пошуку від розміру масиву.

```python
import matplotlib.pyplot as plt

sizes = [1000, 5000, 10000, 50000, 100000, 500000, 1000000]
df_lin = get_linear_search_df(sizes)
df_bin = get_bin_search_df(sizes)

plt.plot(df_lin['n'], df_lin['time'], label='Linear Search')
plt.plot(df_bin['n'], df_bin['time'], label='Binary Search')
plt.xlabel('n')
plt.ylabel('Time (s)')
plt.legend()
plt.show()
```

###**Завдання 5:** Оцінити, який з двох алгоритмів є ефективнішим і в якому діапазоні розміру задачі.

Бінарний пошук значно ефективніший на будь-яких великих масивах ($n > 100$), оскільки його час зростає логарифмічно, тоді як лінійний пошук демонструє пряму лінійну залежність $O(n)$. Лінійний пошук доцільний лише для дуже малих невідсортованих списків, де витрати на сортування (необхідне для бінарного пошуку) перевищують вигоду від швидкості самого пошуку.

###Контрольні питання:
- Задача пошуку полягає у визначенні місцезнаходження (індексу) елемента із заданим значенням (ключем) у певній структурі даних або встановленні факту його відсутності.
- Часова складність: кількість операцій або час, необхідний для пошуку елемента в найгіршому, середньому та найкращому випадках. Просторова складність: обсяг додаткової пам'яті, яку використовує алгоритм під час роботи.
- Лінійний пошук — це найпростіший метод, що полягає у послідовному перегляді всіх елементів масиву від початку до кінця до моменту знаходження потрібного значення.Його асимптотична складність становить $O(n)$.
- Бінарний пошук — це ітераційний алгоритм для відсортованих структур, принцип якого полягає у постійному поділі масиву навпіл: порівнянні цільового значення з центральним елементом і відкиданні тієї половини, де шуканого значення точно немає.
- Бінарний пошук значно швидший на великих даних, але працює виключно з відсортованими масивами. Сортування вимагає додаткового часу, тому для одноразового пошуку в невеликому списку простіше використати лінійний метод. Лінійний пошук універсальний, проте його швидкість падає пропорційно зростанню кількості елементів. Бінарний підхід кращий для стабільних баз даних з частими запитами, а лінійний — для невеликих хаотичних наборів.
- Для прискорення роботи існують стрибковий (Jump Search) та експоненціальний пошуки, які швидше за лінійний знаходять потрібний діапазон елементів. Інтерполяційний пошук ще ефективніший за бінарний на рівномірно розподілених даних, бо він "вгадує" позицію числа. У сучасних системах часто використовують хешування, що дозволяє знаходити елемент миттєво за сталий час $O(1)$. Також існують пошуки в деревах та графах (DFS, BFS) для складних структур даних. Кожен метод обирається залежно від формату зберігання та обсягу інформації.
