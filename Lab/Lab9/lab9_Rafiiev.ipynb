###**Завдання 1:** Тестування алгоритмів 

```python
def naive_match(p, t):
    assert len(p) <= len(t)
    occurrences = []
    for i in range(0, len(t) - len(p) + 1):
        match = True
        for j in range(0, len(p)):
            if t[i+j] != p[j]:
                match = False
                break
        if match:
            occurrences.append(i)
    return occurrences

def z_func(s):
    Z = [len(s)] + [0] * (len(s) - 1)
    if len(s) <= 1:
        return Z
    
    for i in range(1, len(s)):
        if s[i] == s[i - 1]:
            Z[1] += 1
        else:
            break
    
    r, l = 0, 0
    if Z[1] > 0:
        r, l = Z[1], 1
    
    for k in range(2, len(s)):
        if k > r:
            for i in range(k, len(s)):
                if s[i] == s[i - k]:
                    Z[k] += 1
                else:
                    break
            r, l = k + Z[k] - 1, k
        else:
            nbeta = r - k + 1
            Zkp = Z[k - l]
            if nbeta > Zkp:
                Z[k] = Zkp
            else:
                nmatch = 0
                for i in range(r + 1, len(s)):
                    if s[i] == s[i - k]:
                        nmatch += 1
                    else:
                        break
                l, r = k, r + nmatch
                Z[k] = r - k + 1
    return Z

def zMatch(p, t):
    s = p + "$" + t
    Z = z_func(s)
    occurrences = []
    for i in range(len(p) + 1, len(s)):
        if Z[i] == len(p):
            occurrences.append(i - (len(p) + 1))
    return occurrences

def compress_with_z(s):
    z_vec = z_func(s)
    for i in range(1, len(s)):
        if (i + z_vec[i] == len(s)) and (z_vec[i] % i == 0):
            return s[:i]
    return s

# --- Тестування ---
test_text = "banana-ananas-and-panama"
test_pattern = "an"

print("--- Тест 1: Наївний пошук ---")
print(f"Збіги для '{test_pattern}' у '{test_text}':", naive_match(test_pattern, test_text))

print("\n--- Тест 2: Z-функція ---")
word = "abcabcabc"
print(f"Z-функція для '{word}':", z_func(word))

print("\n--- Тест 3: Пошук підрядка через Z-функцію ---")
print(f"Збіги для '{test_pattern}' у '{test_text}':", zMatch(test_pattern, test_text))

print("\n--- Тест 4: Стиснення рядка ---")
periodic_str = "xyzxyzxyzxyz"
print(f"Стиснення '{periodic_str}':", compress_with_z(periodic_str))
print(f"Стиснення 'hello':", compress_with_z("hello"))
```

###**Завдання 2:** Асимптотика наївного пошуку  
**Формула:** $O(|t| \cdot |p|)$.  
**Чому:** Зовнішній цикл зміщує шаблон по тексту ($O(|t|)$), а внутрішній у найгіршому випадку перевіряє всі символи шаблону ($O(|p|)$). Найгірший випадок — повторення однакових символів (наприклад, $t = \text{"AAAA"}$, $p = \text{"AA"}$).

###**Завдання 3:** Асимптотика стиснення через Z-функцію   
**Формула:** $O(n)$, де $n$ — довжина рядка.  
**Чому:** Обчислення Z-функції працює за лінійний час $O(n)$ завдяки скануванню рядка в один бік через «вікно». Фінальний перебір варіантів періоду робить максимум $n$ кроків з операціями за $O(1)$.

###Контрольні питання:
1. Головна відмінність між цими структурами полягає в напрямку аналізу тексту: префікс-функція ($\pi$) для кожної позиції обчислює довжину найдовшого префікса, який одночасно є суфіксом поточного підрядка, тобто аналізує зв'язки «назад», тоді як Z-функція рахує довжину підрядка, що починається з поточної позиції і повністю збігається з початком усього рядка, тобто заглядає «вперед».
2. У вищому контексті Z-функція є потужним інструментом лінійного аналізу конфігурації тексту, значення якого полягає в можливості всього за $O(n)$ часу без побудови складних скінченних автоматів розв'язувати задачі точного пошуку підрядків, швидкого визначення періодичності рядка для його стиснення, а також знаходження кількості унікальних підтекстів і повторів.
3. Для розв'язання цієї задачі використовують три основні підходи: динамічне програмування зі складністю $O(|A| \cdot |B|)$ через побудову матриці збігів символів, високоефективні суфіксні дерева або суфіксні автомати, які знаходять відповідь за лінійний час $O(|A| + |B|)$, а також метод Рабіна-Карпа, що поєднує поліноміальне хешування з бінарним пошуком по довжині підрядка за $O(n \log n)$.
4. Ці алгоритми є фундаментом для прикладних систем NLP та текстового аналізу, де вони застосовуються в пошукових рушіях для миттєвої індексації ключових слів, у модулях автовиправлення, Т9 та перевірки орфографії через метрики схожості на зразок відстані Левенштейна, а також у біоінформатиці для вирівнювання, аналізу та пошуку мутацій у гігантських ланцюжках ДНК і РНК.
