##**Завдання 1:** Виконати теоретичні розрахунки асимптотичної складності алгоритмів сортування методами вставляння та бульбашки.

**Алгоритм сортування бульбашкою** *(Bubble Sort)* базується на ітераційному порівнянні сусідніх елементів. У найгіршому та середньому випадках алгоритм виконує $n-1$ проходів по масиву, де на кожному етапі кількість порівнянь зменшується. Математично це виражається сумою арифметичної прогресії порівнянь, що призводить до квадратичної залежності від кількості елементів. Таким чином, асимптотична складність становить $O(n^2)$. Хоча алгоритм простий у реалізації, він є найменш ефективним для великих масивів через величезну кількість операцій перестановки *(swap)*.  
**Алгоритм сортування вставлянням** *(Insertion Sort)* працює за принципом розділення масиву на відсортовану та невідсортовану частини. На кожному кроці елемент із невідсортованої частини вилучається та вставляється на відповідну позицію у відсортовану, зсуваючи всі більші елементи вправо. У найгіршому випадку (коли масив відсортований у зворотному порядку) складність також становить $O(n^2)$. Проте в "найкращому" випадку (вже відсортований масив) алгоритм робить лише $n-1$ порівняння, що дає лінійну складність $O(n)$.

##**Завдання 2:** Реалізувати на Python алгоритм сортування методом бульбашки.

```python
def bubble_sort(arr):
    n = len(arr)
    
    for i in range(n):
        swapped = False

        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        if not swapped:
            break
    
    return arr

if __name__ == "__main__":
    test_data = [64, 34, 25, 12, 22, 11, 90]
    print(f"Початковий масив: {test_data}")
    
    sorted_data = bubble_sort(test_data)
    print(f"Відсортований масив: {sorted_data}")
```

##**Завдання 3:** Дослідити реальний час виконання обох алгоритмів на Python і побудувати графіки залежностей 𝑇𝑖𝑛𝑠𝑒𝑟𝑡(𝑛) та 𝑇𝑏𝑢𝑏𝑙𝑒(𝑛) для 𝑛 = [5, 10,50,100,500,1000,2000,3000,4000,5000,10000,20000,50000,10000], згенерувавши відповідні цілочисельні масиви.

```python
import time
import random
import matplotlib.pyplot as plt

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        if not swapped:
            break

def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

n_values = [5, 10, 50, 100, 500, 1000, 2000, 3000, 4000, 5000]
t_bubble = []
t_insert = []

for n in n_values:
    original_arr = [random.randint(0, 100000) for _ in range(n)]

    arr_b = original_arr.copy()
    start = time.perf_counter()
    bubble_sort(arr_b)
    t_bubble.append(time.perf_counter() - start)
 
    arr_i = original_arr.copy()
    start = time.perf_counter()
    insertion_sort(arr_i)
    t_insert.append(time.perf_counter() - start)

plt.figure(figsize=(10, 6))
plt.plot(n_values, t_bubble, label='Bubble Sort')
plt.plot(n_values, t_insert, label='Insertion Sort')
plt.xlabel('Розмір масиву (n)')
plt.ylabel('Час виконання (секунди)')
plt.title('Порівняння реального часу виконання алгоритмів')
plt.legend()
plt.grid(True)
plt.show()
```

###**Висновок:** під час виконання цієї лабораторної роботи я засвоїв аналіз складності алгоритмів як технологію на прикладі алгоритмів сортування методами включення та обміну.
##Контрольні питання:
 - O-нотація описує верхню межу складності (найгірший випадок), тобто як швидко росте час виконання. Ω-нотація описує нижню межу (найкращий випадок), тобто мінімальний час виконання.
 - Складність дорівнює O(n), бо масив вже відсортований і кожен елемент лише один раз порівнюється без зсувів.
 - Ефективний алгоритм — це алгоритм, який використовує мінімум часу та пам’яті для розв’язання задачі.
 - Розмір задачі — це кількісна характеристика вхідних даних (наприклад, кількість елементів у масиві), від якої залежить складність алгоритму.
 - Асимптотична складність: O(N^3), бо при великих N домінує старший член N^3.
