###**Завдання 1:** Бінарне дерево (реалізація списками)  
Для видалення гілки ми створимо функцію, яка знаходить піддерево за вказаним шляхом і замінює його на порожній список [].

```python
def BinaryTree(r):
    return [r, [], []]

def insertLeft(root, newBranch):
    t = root.pop(1)
    if len(t) > 1:
        root.insert(1, [newBranch, t, []])
    else:
        root.insert(1, [newBranch, [], []])
    return root

def insertRight(root, newBranch):
    t = root.pop(2)
    if len(t) > 1:
        root.insert(2, [newBranch, [], t])
    else:
        root.insert(2, [newBranch, [], []])
    return root

def deleteBranch(root, side):
    """
    side: 1 для лівої гілки, 2 для правої
    """
    if side == 1 or side == 2:
        root[side] = []
    return root

tree = BinaryTree('A')
insertLeft(tree, 'B')
insertRight(tree, 'C')
insertLeft(tree[1], 'D')

print("Дерево до видалення:", tree)
deleteBranch(tree[1], 1)
print("Дерево після видалення гілки 'D':", tree)
```

####**Завдання 1.1:**Асимптотична складність (Бінарне дерево пошуку / Звичайне дерево)  
У випадку звичайної спискової реалізації або незбалансованого дерева пошуку:
 - Пошук (Search):
   - В середньому: $\Theta(\log n)$, якщо дерево збалансоване (як BST). Проте для звичайного дерева без порядку — це $\Theta(n)$, оскільки треба перевірити кожен вузол.
   - Найгірший випадок: $O(n)$. Це стається, коли дерево вироджується у «бамбук» (стає схожим на список).
 - Вставка (Insert):
   - В середньому: $\Theta(1)$ — якщо ми вже знаємо батьківський вузол і просто додаємо нащадка. $\Theta(\log n)$ — якщо треба знайти місце для вставки в сортованому дереві.
   - Найгірший випадок: $O(n)$, якщо для пошуку місця вставки доводиться проходити через усі рівні виродженого дерева.
 - Видалення (Delete):
   - В середньому: $\Theta(1)$ для видалення гілки за посиланням.
   - Найгірший випадок: $O(n)$, якщо операція передбачає попередній пошук елемента, який потрібно видалити.

###**Завдання 2:** Завдання 2: Купа (Heap)
Для генерації купи з довільного масиву використовується метод «занурення» (down-heap) для всіх нелистових вузлів.

```python
import random

heap = [0]
currSize = 0

def parent(i):
    return i // 2

def left(i):
    return 2 * i

def right(i):
    return 2 * i + 1

def swap(a, b):
    heap[a], heap[b] = heap[b], heap[a]

def maxHeapify(index):
    """Підтримує властивість max-heap, опускаючи елемент вниз"""
    global currSize
    lar = index
    l = left(index)
    r = right(index)

    if l <= currSize and heap[l] > heap[lar]:
        lar = l
    if r <= currSize and heap[r] > heap[lar]:
        lar = r
        
    if lar != index:
        swap(index, lar)
        maxHeapify(lar)

def buildHeap(alist):
    """Створює купу з довільного масиву за час O(n)"""
    global currSize, heap
    currSize = len(alist)
    heap = [0] + alist[:]
    i = len(alist) // 2
    while i > 0:
        maxHeapify(i)
        i -= 1

def insert(elem):
    """Додає новий елемент у купу та піднімає його вгору"""
    global currSize
    heap.append(elem)
    currSize += 1
    index = currSize
    par = parent(index)
    
    while index > 1 and heap[par] < heap[index]:
        swap(par, index)
        index = par
        par = parent(index)

def extractMax():
    """Вилучає та повертає корінь (максимум), відновлюючи купу"""
    global currSize
    if currSize < 1:
        return None
    
    maxElem = heap[1]
    heap[1] = heap[currSize]
    heap.pop()
    currSize -= 1
    
    if currSize > 0:
        maxHeapify(1)
    return maxElem

random_list = [random.randint(1, 100) for _ in range(10)]
print(f"Початковий масив: {random_list}")

buildHeap(random_list)
print(f"Купа після buildHeap: {heap[1:]}")

my_number = 22
insert(my_number)
print(f"Купа після додавання номера {my_number}: {heap[1:]}")

max_val = extractMax()
print(f"Вилучено максимальний елемент: {max_val}")
print(f"Фінальний стан купи: {heap[1:]}")
```

####**Завдання 2.1:** Асимптотична складність купа (Binary Heap)  
Купа завжди підтримує властивість повноти, тому її висота суворо обмежена логарифмом.
  - Пошук (Search):  
    - Максимальний/Мінімальний елемент: $\Theta(1)$, оскільки він завжди знаходиться в корені (перший елемент масиву).  
    - Довільний елемент: $O(n)$, тому що купа не впорядкована для швидкого пошуку будь-якого значення (лівий нащадок не обов'язково менший за правого).  
  -Вставка (Insert):  
    - В середньому та найгіршому: $O(\log n)$.   
    - Новий елемент додається в кінець масиву і «спливає» вгору до своєї позиції. Кількість обмінів не перевищує висоту дерева.  
  - Видалення максимуму (Extract Max):  
    - В середньому та найгіршому: $O(\log n)$.   
    - Корінь замінюється останнім елементом, який потім «занурюється» вниз (MaxHeapify).  

###**Завдання 3:** Геш-таблиця та колізії
1. Методи розв'язання колізій
   - Метод ланцюжків (Chaining): Кожна комірка таблиці містить список (ланцюжок) усіх елементів, які мають однаковий хеш-індекс.
   - Відкрита адресація: При виникненні колізії шукається інша вільна комірка за певним алгоритмом (лінійне пробування, квадратичне пробування або подвійне хешування).
2. Реалізація геш-таблиці з ланцюжками

```python
import time

class ChainHashTable:
    def __init__(self, size):
        self.size = size
        self.table = [[] for _ in range(self.size)]

    def _hash(self, key):
        return hash(key) % self.size

    def insert(self, key, value):
        index = self._hash(key)
        for kvp in self.table[index]:
            if kvp[0] == key:
                kvp[1] = value
                return
        self.table[index].append([key, value])

    def search(self, key):
        index = self._hash(key)
        for kvp in self.table[index]:
            if kvp[0] == key: return kvp[1]
        return None

    def delete(self, key):
        index = self._hash(key)
        for i, kvp in enumerate(self.table[index]):
            if kvp[0] == key:
                self.table[index].pop(i)
                return True
        return False

ht = ChainHashTable(100)
test_data = [
    (123, "Integer"),
    ("Python", "String"),
    ((1, 2), "Tuple")
]

for k, v in test_data:
    start = time.perf_counter()
    ht.insert(k, v)
    end = time.perf_counter()
    print(f"Тип {type(k).__name__}: час вставки {end-start:.8f}с")
```

Висновки щодо типів даних
Працездатність: Геш-таблиця працює з будь-якими незмінними (immutable) типами даних (int, str, tuple). Списки (list) та словники (dict) не можуть бути ключами, оскільки вони не мають сталого хеш-значення.
Швидкість: Для більшості типів час вставки та пошуку є константним O(1). Проте для дуже довгих рядків або складних кортежів час обчислення хеш-функції може незначно зростати.

 ####**Завдання 3.1:** Геш-таблиця (Hash Table)
 Ефективність тут прямо залежить від якості геш-функції та коефіцієнта заповнення.
  - Пошук (Search):
    - В середньому: $\Theta(1)$. Ідеальна ситуація, коли ми отримуємо прямий доступ до комірки масиву за індексом.
    - Найгірший випадок: $O(n)$. Це трапляється при виникненні великої кількості колізій, коли всі елементи потрапляють в один «ланцюжок» або збиваються в один кластер.
  - Вставка (Insert):
    - В середньому: $\Theta(1)$. Обчислюється геш і елемент додається в таблицю.
    - Найгірший випадок: $O(n)$. Окрім колізій, може знадобитися операція rehashing (перебудова всієї таблиці при збільшенні її розміру).
  - Видалення (Delete):
    - В середньому: $\Theta(1)$. Швидкий доступ до елемента за ключем.Найгірший випадок: $O(n)$.
    - Аналогічно пошуку — через необхідність проходити довгий список колізій.

###Контрольні питання:
- **Бінарне дерево vs BST:** звичайне бінарне дерево не має правил розташування значень — вузол просто має до двох нащадків. Бінарне дерево пошуку (BST) суворо впорядковане: лівий нащадок завжди менший за батька, а правий — більший, що дозволяє шукати дані за $O(\log n)$.
- **Бінарне дерево vs Купа:** купа — це завжди повне дерево (заповнене без «дірок»), де головне — пріоритет. У купі батько завжди більший (або менший) за дітей, але немає різниці між лівою та правою стороною, на відміну від BST.
- **Типи дерев:** крім базових, існують збалансовані дерева (AVL, Червоно-чорні), які не дають дереву стати занадто глибоким, та B-дерева, що використовуються в базах даних для роботи з величезними обсягами даних на дисках.
- **Застосування дерев:** незамінні для представлення ієрархій (файлові системи, HTML-документи), швидкого пошуку в базах даних, маршрутизації мережевого трафіку та стиснення файлів (дерево Хаффмана).
- **Організація купи:** купа зберігається як звичайний масив. При додаванні новий елемент ставиться в кінець і «спливає» вгору, доки не знайде своє місце. При вилученні корінь замінюється останнім елементом, який потім «занурюється» вниз для відновлення порядку.
- **Застосування купи:** найефективніша для черг з пріоритетами (наприклад, обробка системних процесів), сортування масивів (Heapsort) та в алгоритмах пошуку найкоротшого шляху на графах.
- **Геш-таблиця:** використовує геш-функцію, щоб перетворити будь-який ключ (текст чи число) в індекс масиву. Це дозволяє знаходити дані практично миттєво ($O(1)$), бо ми одразу знаємо номер комірки, де лежить об'єкт.
- **Вирішення колізій:** Коли два ключі мають однаковий індекс, використовують метод ланцюжків (створення списку в комірці) або відкриту адресацію (пошук наступної вільної комірки). Ланцюжки надійніші, але відкрита адресація економить пам'ять у невеликих таблицях.
